```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

time = ctrl.Antecedent(np.arange(10,120,1),'time')
rice = ctrl.Antecedent(np.arange(100,1000,1),'rice')
power  = ctrl.Antecedent(np.arange(0,100,1),'power')

#time
time ['Rất nhanh']=fuzz.trimf(time .universe,[10,10,20])
time ['Khá nhanh']=fuzz.trimf(time .universe,[10,20,50])
time ['Nhanh']=fuzz.trimf(time .universe,[20,50,75])
time ['Bình thường']=fuzz.trimf(time .universe,[50,75,100])
time ['Chậm']=fuzz.trimf(time .universe,[75,100,120])
time ['Rất chậm']=fuzz.trimf(time .universe,[100,120,120])

#rice
rice['Rất ít']=fuzz.trimf(rice.universe,[100,100,280])
rice['Khá ít']=fuzz.trimf(rice.universe,[100,280,460])
rice['Ít']=fuzz.trimf(rice.universe,[280,460,640])
rice['Bình thường']=fuzz.trimf(rice.universe,[460,640,820])
rice['Nhiều']=fuzz.trimf(rice.universe,[640,820,1000])
rice['Rất nhiều']=fuzz.trimf(rice.universe,[820,1000,1000])

#power
power['Rất nhẹ']=fuzz.trimf(power.universe,[0,0,20])
power['Nhẹ']=fuzz.trimf(power.universe,[0,20,40])
power['Bình thường']=fuzz.trimf(power.universe,[20,40,60])
power['Mạnh']=fuzz.trimf(power.universe,[40,60,80])
power['Khá mạnh']=fuzz.trimf(power.universe,[60,80,100])
power['Rất mạnh']=fuzz.trimf(power.universe,[80,100,100])

time.view()
rice.view()
power.view()

rule1=ctrl.Rule(time['Rất nhanh'] & rice['Rất ít'], power['Rất nhẹ'])
rule2=ctrl.Rule(time['Rất nhanh'] & rice['Khá ít'], power['Nhẹ'])
rule3=ctrl.Rule(time['Rất nhanh'] & rice['Ít'], power['Bình thường'])
rule4=ctrl.Rule(time['Rất nhanh'] & rice['Bình thường'], power['Mạnh'])
rule5=ctrl.Rule(time['Rất nhanh'] & rice['Nhiều'], power['Khá mạnh'])
rule6=ctrl.Rule(time['Rất nhanh'] & rice['Rất nhiều'], power['Rất mạnh']) 

rule7=ctrl.Rule(time['Khá nhanh'] & rice['Rất ít'], power['Rất nhẹ'])
rule8=ctrl.Rule(time['Khá nhanh'] & rice['Khá ít'], power['Nhẹ'])
rule9=ctrl.Rule(time['Khá nhanh'] & rice['Ít'], power['Bình thường'])
rule10=ctrl.Rule(time['Khá nhanh'] & rice['Bình thường'], power['Mạnh'])
rule11=ctrl.Rule(time['Khá nhanh'] & rice['Nhiều'], power['Khá mạnh'])
rule12=ctrl.Rule(time['Khá nhanh'] & rice['Rất nhiều'], power['Rất mạnh']) 

rule13=ctrl.Rule(time['Nhanh'] & rice['Rất ít'], power['Rất nhẹ'])
rule14=ctrl.Rule(time['Nhanh'] & rice['Khá ít'], power['Nhẹ'])
rule15=ctrl.Rule(time['Nhanh'] & rice['Ít'], power['Bình thường'])
rule16=ctrl.Rule(time['Nhanh'] & rice['Bình thường'], power['Mạnh'])
rule17=ctrl.Rule(time['Nhanh'] & rice['Nhiều'], power['Khá mạnh'])
rule18=ctrl.Rule(time['Nhanh'] & rice['Rất nhiều'], power['Rất mạnh']) 

rule19=ctrl.Rule(time['Bình thường'] & rice['Rất ít'], power['Rất nhẹ'])
rule20=ctrl.Rule(time['Bình thường'] & rice['Khá ít'], power['Nhẹ'])
rule21=ctrl.Rule(time['Bình thường'] & rice['Ít'], power['Bình thường'])
rule22=ctrl.Rule(time['Bình thường'] & rice['Bình thường'], power['Mạnh'])
rule23=ctrl.Rule(time['Bình thường'] & rice['Nhiều'], power['Khá mạnh'])
rule24=ctrl.Rule(time['Bình thường'] & rice['Rất nhiều'], power['Rất mạnh']) 

rule25=ctrl.Rule(time['Chậm'] & rice['Rất ít'], power['Rất nhẹ'])
rule26=ctrl.Rule(time['Chậm'] & rice['Khá ít'], power['Nhẹ'])
rule27=ctrl.Rule(time['Chậm'] & rice['Ít'], power['Bình thường'])
rule28=ctrl.Rule(time['Chậm'] & rice['Bình thường'], power['Mạnh'])
rule29=ctrl.Rule(time['Chậm'] & rice['Nhiều'], power['Khá mạnh'])
rule30=ctrl.Rule(time['Chậm'] & rice['Rất nhiều'], power['Rất mạnh']) 

rule31=ctrl.Rule(time['Rất chậm'] & rice['Rất ít'], power['Rất nhẹ'])
rule32=ctrl.Rule(time['Rất chậm'] & rice['Khá ít'], power['Nhẹ'])
rule33=ctrl.Rule(time['Rất chậm'] & rice['Ít'], power['Bình thường'])
rule34=ctrl.Rule(time['Rất chậm'] & rice['Bình thường'], power['Mạnh'])
rule35=ctrl.Rule(time['Rất chậm'] & rice['Nhiều'], power['Khá mạnh'])
rule36=ctrl.Rule(time['Rất chậm'] & rice['Rất nhiều'], power['Rất mạnh']) 

powering_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,
rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25,rule26,rule27,rule28,rule29,rule30,rule31,rule32,rule33,rule34,rule35,rule36])
powering= ctrl.ControlSystemSimulation(powering_ctrl)
powering.input['time']=75
powering.input['rice']=640

powering.compute()

print(powering.output['power'])
power.view(sim=powering)
```
