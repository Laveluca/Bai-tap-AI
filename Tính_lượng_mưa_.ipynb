```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
wind = ctrl.Antecedent(np.arange(0.7,3.1,0.1),'wind')
rain = ctrl.Antecedent(np.arange(0.2,9.1,0.1),'rain')
temp = ctrl.Antecedent(np.arange(20.4,41.2,0.1),'temp')

#wind
wind['Very Low']=fuzz.trimf(wind.universe,[0.7,0.7,1.3])
wind['Low']=fuzz.trimf(wind.universe,[0.7,1.3,1.9])
wind['Normal']=fuzz.trimf(wind.universe,[1.3,1.9,2.5])
wind['Hard']=fuzz.trimf(wind.universe,[1.9,2.5,3.1])
wind['Very Hard']=fuzz.trimf(wind.universe,[2.5,3.1,3.1])

#Temperature
temp['Very Low']=fuzz.trimf(temp.universe,[20.4,20.4,25.6])
temp['Low']=fuzz.trimf(temp.universe,[20.4,25.6,30.8])
temp['Normal']=fuzz.trimf(temp.universe,[25.6,30.8,36])
temp['Hard']=fuzz.trimf(temp.universe,[30.8,36,41.2])
temp['Very Hard']=fuzz.trimf(temp.universe,[36,41.2,41.2])

#rain
rain['Very Low']=fuzz.trimf(rain.universe,[0.2,1.68,3.17])
rain['Low']=fuzz.trimf(rain.universe,[1.68,3.17,4.65])
rain['Normal']=fuzz.trimf(rain.universe,[3.17,4.65,6.13])
rain['Hard']=fuzz.trimf(rain.universe,[4.65,6.13,7.61])
rain['Very Hard']=fuzz.trimf(rain.universe,[6.13,7.61,9.1])

wind.view()
temp.view()
rain.view()

rule1=ctrl.Rule(wind['Very Low'] and temp['Very Low'],rain['Very Low'])
rule2=ctrl.Rule(wind['Low']and temp['Very Low'],rain['Very Low'])
rule3=ctrl.Rule(wind['Normal']and temp['Very Low'],rain['Low'])
rule4=ctrl.Rule(wind['Hard']and temp['Very Low'],rain['Hard'])
rule5=ctrl.Rule(wind['Very Hard']and temp['Very Low'],rain['Very Hard'])

rule6=ctrl.Rule(wind['Very Low']and temp['Low'],rain['Very Low'])
rule7=ctrl.Rule(wind['Low']and temp['Low'],rain['Very Low'])
rule8=ctrl.Rule(wind['Normal']and temp['Low'],rain['Low'])
rule9=ctrl.Rule(wind['Hard']and temp['Low'],rain['Low'])
rule10=ctrl.Rule(wind['Very Hard']and temp['Low'],rain['Hard'])

rule11=ctrl.Rule(wind['Very Low']and temp['Normal'],rain['Low'])
rule12=ctrl.Rule(wind['Low']and temp['Normal'],rain['Low'])
rule13=ctrl.Rule(wind['Normal']and temp['Normal'],rain['Low'])
rule14=ctrl.Rule(wind['Hard']and temp['Normal'],rain['Hard'])
rule15=ctrl.Rule(wind['Very Hard']and temp['Normal'],rain['Hard'])

rule16=ctrl.Rule(wind['Very Low']and temp['Hard'],rain['Low'])
rule17=ctrl.Rule(wind['Low']and temp['Hard'],rain['Hard'])
rule18=ctrl.Rule(wind['Normal']and temp['Hard'],rain['Hard'])
rule19=ctrl.Rule(wind['Hard']and temp['Hard'],rain['Very Hard'])
rule20=ctrl.Rule(wind['Very Hard']and temp['Hard'],rain['Hard'])

rule21=ctrl.Rule(wind['Very Low']and temp['Very Hard'],rain['Low'])
rule22=ctrl.Rule(wind['Low']and temp['Very Hard'],rain['Hard'])
rule23=ctrl.Rule(wind['Normal']and temp['Very Hard'],rain['Very Hard'])
rule24=ctrl.Rule(wind['Hard']and temp['Very Hard'],rain['Very Hard'])
rule25=ctrl.Rule(wind['Very Hard']and temp['Very Hard'],rain['Very Hard'])

raining_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
raining= ctrl.ControlSystemSimulation(raining_ctrl)
raining.input['wind']=3
raining.input['temp']=30
raining.compute()
print(raining.output['rain'])
rain.view(sim=raining)

```
